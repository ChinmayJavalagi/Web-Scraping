```python
#importing libraries

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
#reading thr URL

http_object = urlopen("https://www.flipkart.com/q/best-laptops-under-rs-50000")
webdata = http_object.read()
print(webdata)
```

```python
soupdata = soup(webdata)
```

```python
dir(soupdata)
```

```python
containers = soupdata.findAll('div', {'class':'_2kHMtA'})
print(type(containers),len(containers))
```
